This project helps find what factors have the most affect on the price of a house in the state of California. We will solve this problems using pandas, numpy,matplotlib and scikit-learn's Linear Regression model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#convert the data from a csv file into a dataframe so that it can be used to experiment with
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df
```

Let's plot each column against the median_house_value column to see which one has the greatest affect on the house value. **Longitue and Latitude will not be used for obvious reasons**

```python
plt.scatter(df['housing_median_age'], df['median_house_value'])
plt.show()
```

```python
plt.scatter(df['total_rooms'], df['median_house_value'])
plt.show()
```

```python
plt.scatter(df['total_bedrooms'], df['median_house_value'])
plt.show()
```

```python
plt.scatter(df['population'], df['median_house_value'])
plt.show()
```

```python
plt.scatter(df['households'], df['median_house_value'])
plt.show()
```

```python
plt.scatter(df['median_income'], df['median_house_value'])
plt.show()
```

Based on the graphs the columns that have the greatest affect on the price of a house are the following: median age and median income. Let's test to see if this is really true

```python
X = df[['housing_median_age', 'median_income']]
y = df['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)

plt.scatter(y_test, pred)
plt.show()
```

As you can see, our hypothesis was, for the most part, correct. The median house age and the median income were the two greatest factors in determing the price of a house in California
